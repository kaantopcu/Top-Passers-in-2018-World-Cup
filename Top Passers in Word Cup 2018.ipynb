```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading all passing data from StatsBomb
allPasses = pd.read_csv("AllPasses.csv")
allPasses
```

```python
# Listing the columns of allPasses
allPasses.columns
```

```python
# Player id and names
playersList = allPasses[['player.id','player.name']].drop_duplicates().reset_index(drop = True)
playersList
```

```python
# Team id and names
teamsList = allPasses[['team.id','team.name']].drop_duplicates().reset_index(drop = True)
teamsList
```

```python
# Total number of passes for each player in each match
totalPasses = allPasses.groupby(['match_id','player.id','team.id'])['type.name'].count().reset_index()
totalPasses.rename(columns={"type.name": "TotalPasses"}, inplace = True)
totalPasses
```

```python
# Unsuccessfull passes dataframe
unsuccessfullPasses = allPasses[allPasses['pass.outcome.name'].notnull()]
unsuccessfullPasses = unsuccessfullPasses.groupby(['match_id','player.id','team.id'])['type.name'].count().reset_index()
unsuccessfullPasses.rename(columns={"type.name": "UnsuccessfullPasses"}, inplace = True)
unsuccessfullPasses
```

```python
# Total Stats for each match
matches = pd.merge(totalPasses,unsuccessfullPasses,how = 'left', on = ['player.id','match_id','team.id'])
matches = matches.fillna(0)
matches['SuccessfullPasses'] = matches['TotalPasses'] - matches['UnsuccessfullPasses']
matches
```

```python
# Merging dataframes for player and team name
matches = pd.merge(playersList,matches,on = ['player.id'])
matches = pd.merge(teamsList,matches,on = ['team.id'])
matches
```

```python
# Teams total stats
teamStats = matches.groupby(['team.name'])['SuccessfullPasses','TotalPasses'].sum().reset_index()
teamStats['SuccessRate'] = teamStats['SuccessfullPasses'] / teamStats['TotalPasses']
teamStats = teamStats.sort_values(by = ['SuccessRate','TotalPasses'],ascending = False).reset_index(drop=True)

# Add passing completion rankings as a new column
teamStats.insert(0, 'Passing Ranking', range(1, 1 + len(teamStats)))

# Adding competition ranking as a column
teamStats['Competition Ranking'] = ""
teamStats.iloc[[0,6,7,9],5] = "Eliminated in Round of 16"
teamStats.iloc[[1,3,8],5] = "Eliminated in Group Phases"
teamStats.iloc[[2,5],5] = "Eliminated in Semi-Finals"
teamStats.iloc[4,5] = "Eliminated in Quarter-Finals"

teamStats.head(10)
```

```python
# Players total stats
playersStats = matches.groupby(['player.name','team.name'])['SuccessfullPasses','TotalPasses'].sum().reset_index()
playersStats['SuccessRate'] = playersStats['SuccessfullPasses'] / playersStats['TotalPasses']

# Players with at least 50 passes are filtered 
playersStats = playersStats[playersStats.TotalPasses >= 100]
playersStats.sort_values(by = ['SuccessRate','TotalPasses'],ascending = False).head(10)
```

```python
# Stats of each teams in each matches
teamMatches = matches.groupby(['team.name','match_id'])['SuccessfullPasses','TotalPasses'].sum().reset_index()
teamMatches['SuccessRate'] = teamMatches['SuccessfullPasses'] / teamStats['TotalPasses']
teamMatches.sort_values(by = ['SuccessRate','TotalPasses'],ascending = False).head(10)
```

```python
# Stats of each players in each matches
playerMatches = matches.groupby(['player.name','team.name','match_id'])['SuccessfullPasses','TotalPasses'].sum().reset_index()
playerMatches['SuccessRate'] = playerMatches['SuccessfullPasses'] / playerMatches['TotalPasses']

# Players with at least 35 passes are filtered 
playerMatches = playerMatches[playerMatches.TotalPasses >= 35]

playerMatches['Opponent Team'] = ""
playerMatches = playerMatches.sort_values(by = ['SuccessRate','TotalPasses'],ascending = False).head(10)
```
